```python
"""Burada ise her 4 bazı da içeren 8 bazdan oluşan genler oluşturdum random bir şekilde. yani her bir sekizli bir geni temsil ediyor
  indisine göre random genler elde etmiş oldum.Daha sonra bu oluşturduğum genler için indisinin 4 e göre modunu aldım yani 4 bazımız
  var ve her 4 bazda bir tekrar etsin istedim( her bir sekizlinin ilk indisi için söylüyorum).Daha sonra bu genlere
  random hastalıklar ekledim.Random sayılar vererek...daha sonra bu genlerde hastalık 0 ise sağlıklı; daha küçük olasılıklarla
  1 ise otizm; 2 ise kas yetersizliği;
  3 ise diğer başka bir hastalık dedim ve o gene o hastalığı ekledim. BU ÖRNEK ÇALIŞMAM!"""


from random import choice, random, randint
allgenes = set()
genes = ""
nucleotides = ['A', 'T' ,'G', 'C']
indices = [0, 0, 0, 0, 0, 0, 0, 0]
for i in range(int(pow(4,8))):
    gene = ""
    for index in indices:
        gene+=nucleotides[index]
    for index in range(len(indices)):
        if i % int(pow(4,index)) == 0:
            indices[index] = (indices[index]+1)%4
    if gene in allgenes:
        continue
    allgenes.add(gene)
    disease = 0 #Healthy
    chance = random()
    if chance > 0.9999:
        disease = 1 #Autism
    elif chance > 0.9998:
        disease = 2 #Muscle Deficiency
    elif chance > 0.9997:
        disease = 3 #Another illness

    genes += gene + " " + str(disease)+"\n"
```

```python
healthy_genes = []
autism_genes = []
muscle_deficiency_genes = []
disease_genes = []
"""burada sağlıklı genler; otizmli genler ; kas yetersizliği genleri ve diğer hastalıklı genler için listeler oluşturdum"""


lines = genes.split("\n")
for line in lines:
    if line == "":
        continue
    gene = line.split()
    if gene[1] == "0":
        healthy_genes.append(gene[0])

    elif gene[1] == "1":
        autism_genes.append(gene[0])

    elif gene[1] == "2":
        muscle_deficiency_genes.append(gene[0])

    elif gene[1] == "3":
        disease_genes.append(gene[0])

"""sonra oluşturduğum her 8 elemanlı genimi satır satır yazdırıyorum ve yanlarındaki boşluktan sonraki ikinci elemana rand sayılar
yazdırmıştım .İşte eğer geni boşluktan böldüğümde ikinci elemanı 0 ise o geni sağlıklı genler listesine ekliyorum ; eğer 1 ise otizmli
genler; 2 ise kas yetersizliği genleri; 3 ise diğer hastalıklı genler listesine ekletiyorum"""

def StrandGenerator():
    strand = ""
    isHealthy = random() < 0.80
    if isHealthy:
        for i in range(10000):
            index = randint(0,len(healthy_genes)-1)
            strand += healthy_genes[index]
    else:
        disease = random()
        faulty_gene = ""
        if disease > 0.7:
            faulty_gene = autism_genes[randint(0,len(autism_genes)-1)]

        elif disease > 0.4:
            faulty_gene = muscle_deficiency_genes[randint(0,len(muscle_deficiency_genes)-1)]

        else:
            faulty_gene = disease_genes[randint(0,len(disease_genes)-1)]
        faulty_index = randint(0,9999)

        for i in range(10000):
            index = randint(0,len(healthy_genes)-1)
            if i == faulty_index:
                strand += faulty_gene
            else:
                strand += healthy_genes[index]
    return strand

    """burada bir strand generator tanımlıyorum.İlk olarak random bir şekide 10000 bazdan oluşan DNA ipliği oluşturuyorum.Öncelikle genim
    hastalıklı mı değil mi onu kontrol etmem gerekecekti .Bu yüzden oransal dağılım oluşturdum ve %80 oranla ipliğimdeki genler sağlıklı
    %20 oranla ise hastalıklı olsun dedim.Eğer ki genim sağlıklı ise onu iplikten çeksin ve sağlıklı gen olarak alsın diğer durumda
     % 20 lik dilime düşer yani burada ipliğim bu oranda hastalıklı genlerden oluşuyor ve burada da oransal dağılım kullandım.Burada
     random bir şekilde hastalıklar daha doğrusu hastalıklı genler atadım(bu %20 lik oran içerisine ) hatalı genlerimde; %30 oranında
     otizm geni;%30 oranında kas yetersizliğ geni ve %40 oranda da diğer hastalık geni bulunsun istedim .BU HATALI GENİ OLUŞTURURKEN
     ilk olarak sağlıklı genlerden bir gen çekiyoruz ve 8 bazdan oluştuğu için dolayısıyla random olarak bir indis atıyoruz yani
      sağlık gendeki bir indisi olmaması gereken bir indisle değiştirmiş oluyoruz.Sonra oluşturduğumuz ipliği returnluyoruz."""
```

```python
""" burada ipliği genlere bölmek için böyle bir fonksiyon kullandım .Yani ipliğim 8 bazdan oluşan gen adı verilen parçalara bölünmüş oldu
sonra bu genler yani oluşan genler için gen listesinde geziyoruz eğer o gen iplikte var ise true yoksa false dönüyor yani hastalıklı
 gen varsa true yoksa false döner gibi"""


def SplitStrandByGenes(strand):
        for start in range(0, len(strand), 8):
            yield strand[start:start+8]
def StrandHasGene(gene_list, strand):
    strand_genes = list(SplitStrandByGenes(strand))
    for gene in gene_list:
        if gene in strand_genes:
            return True
    return False
```
